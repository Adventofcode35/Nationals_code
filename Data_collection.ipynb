```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers, models, optimizers, callbacks
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from PIL import Image  # Use PIL to avoid OpenCV dependencies
import matplotlib.pyplot as plt

# 1. Define Data Directory and Classes
data_dir = '/kaggle/working'  # Root directory containing the 7 class folders
class_names = ['Canines', 'Cars', 'Cattles', 'Elephant', 'Trains', 'Humans', 'Trucks']
image_size = (32, 32)  # Target image size for the ESP32-CAM

# 2. Load and Preprocess Data
def load_and_preprocess_data(data_dir, image_size, class_names):
    images = []
    labels = []
    for i, class_name in enumerate(class_names):
        class_dir = os.path.join(data_dir, class_name)
        for image_file in os.listdir(class_dir):
            image_path = os.path.join(class_dir, image_file)
            try:
                image = Image.open(image_path).resize(image_size).convert('RGB') # Ensure RGB for consistent input
                image = np.array(image) / 255.0  # Normalize pixel values
                images.append(image)
                labels.append(i) # Using index of class name for label
            except Exception as e:
                print(f"Error processing {image_path}: {e}")

    return np.array(images), np.array(labels)

images, labels = load_and_preprocess_data(data_dir, image_size, class_names)

# 3. Data Splitting (Train/Validation/Test)
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.33, random_state=42) # 33% test
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)  # 20% validation, 80% train (of remaining)
y_train_encoded = to_categorical(y_train, num_classes=len(class_names))
y_val_encoded = to_categorical(y_val, num_classes=len(class_names))
y_test_encoded = to_categorical(y_test, num_classes=len(class_names))
y_train_encoded = tf.cast(y_train_encoded, tf.float32)
y_val_encoded = tf.cast(y_val_encoded, tf.float32)
y_test_encoded = tf.cast(y_test_encoded, tf.float32)
# 4. Define the CNN Model
def create_cnn_model(input_shape, num_classes):
    model = models.Sequential([
        layers.Conv2D(64, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.2),

        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.3),

        layers.Conv2D(16, (3, 3), activation='relu'),
        layers.Flatten(),

        layers.Dense(num_classes, activation='softmax')  # Softmax for multi-class classification
    ])
    return model

# 5. Create the Model Instance
input_shape = (image_size[0], image_size[1], 3) # 3 channels for RGB images
num_classes = len(class_names)
model = create_cnn_model(input_shape, num_classes)

# 6. Compile the Model
model.compile(optimizer=optimizers.Adam(),
              loss='sparse_categorical_crossentropy', # Use for integer labels
              metrics=['accuracy', 'MeanSquaredError', 'MeanAbsoluteError']) # monitor accuracy, mse, and mae

model.summary()  # Print model summary to see the architecture

# 7. Define Early Stopping Callback
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True) #early stopping callback

# 8. Train the Model
epochs = 30
batch_size = 32

history = model.fit(X_train, y_train_encoded,
                    epochs=epochs,
                    batch_size=batch_size,
                    validation_data=(X_val, y_val_encoded),
                    callbacks=[early_stopping])

# 9. Evaluate the Model on the Test Set
test_loss, test_accuracy, test_mse, test_mae = model.evaluate(X_test, y_test_encoded, verbose=0) #evaluate on test data

print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
print(f"Test MSE: {test_mse}")
print(f"Test MAE: {test_mae}")

# 10. Baseline Model (for comparison) - A simpler model or a random classifier

# Create a random baseline classifier:
from sklearn.dummy import DummyClassifier
dummy_clf = DummyClassifier(strategy="most_frequent")  # Or "uniform", "stratified"

# Train the baseline model (this step is still necessary!)
dummy_clf.fit(X_train, y_train)

# Evaluate on test set:
baseline_accuracy = dummy_clf.score(X_test, y_test)
print(f"Baseline Accuracy: {baseline_accuracy}")


# 11. Plot Training History (Loss and Accuracy)

# Extract training and validation loss values
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# Extract training and validation accuracy values
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

# Create a figure with two subplots (loss and accuracy)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Plot training and validation loss
axes[0].plot(train_loss, label='Training Loss')
axes[0].plot(val_loss, label='Validation Loss')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Training and Validation Loss')
axes[0].legend()

# Plot training and validation accuracy
axes[1].plot(train_accuracy, label='Training Accuracy')
axes[1].plot(val_accuracy, label='Validation Accuracy')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Training and Validation Accuracy')
axes[1].legend()

# Adjust layout and display the plot
plt.tight_layout()
plt.show()

# Save Model (Optional)
model.save('esp32_cam_model.h5') # Save the model for later use with ESP32 CAM (requires conversion)

print("Training complete.")
```
